```python
import seaborn as sns
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
```

```python
model1 = pd.read_excel("./Model1_Output.xlsx")
model1.head()
```

```python
model2 = pd.read_excel("./Model2_Output.xlsx")
model2.head()
```

```python
labels_model1=model1.Actual_class.unique().tolist()
labels_model2=model2.Actual_class.unique().tolist()
```

```python
cm_for_model1 = confusion_matrix(model1.Actual_class, model1.Predicted_class,labels=labels_model1)
cm_for_model2 = confusion_matrix(model2.Actual_class, model2.Predicted_class, labels=labels_model2)
```

```python
fig = plt.figure(figsize=(20,5))
ax1 = plt.subplot(121)
ax1.set_title("Model1")
sns.heatmap(cm_for_model1, annot=True,cmap="Greens",xticklabels=labels_model1, yticklabels=labels_model1)
plt.xlabel('Actual',labelpad=20, color="Blue")
plt.ylabel('Predicted', color="Blue")
ax2 = plt.subplot(122)
ax2.set_title("Model2")
sns.heatmap(cm_for_model2, annot=True,cmap="Greens",xticklabels=labels_model2, yticklabels=labels_model2)
plt.xlabel('Actual',labelpad=20, color="Blue")
plt.ylabel('Predicted',color="Blue")
plt.show()
```

```python
pd.DataFrame(classification_report(model1.Actual_class, model1.Predicted_class,output_dict=True)).T
```

```python
pd.DataFrame(classification_report(model2.Actual_class, model2.Predicted_class,output_dict=True)).T
```
